### Since the sensor data is mostly skewed, null values are replaced with the median.

```python
for column in dt_sensor.columns:
  if dt_sensor[column].isnull().any():
    dt_sensor[column].fillna(dt_sensor[column].median(),inplace = True)
```

```python
dt_sensor.info()
```

```python
dt_sensor = dt_sensor.set_index('timestamp')
dt_sensor.head()
```
